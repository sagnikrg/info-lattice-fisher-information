```julia
#########################
# Include Headers
#########################

using LinearAlgebra
using Arpack

include("../header/.src/circuit/brickwall.jl")
include("../header/.src/info_lattice/info_lattice.jl")
include("../header/.src/plotmods/colorschemes_mods.jl")
include("../header/.src/random_matrices/random_matrices.jl")
```

```julia
## CKT

L=8
epsilon=0.999
theta=0.0

U=brickwall(L,theta, epsilon)

eigvals,eigvecs=eigen(U);
```

```julia
 info_lattice_t= Vector{Vector{Float64}}(undef, L)

 Psi=eigvecs[1,:]
 info_lattice_t.=info_lattice(Psi, L)

 for i in 2:length(eigvals)
    Psi=eigvecs[i,:]
    info_lattice_t+=info_lattice(Psi,L)
 end

 info_lattice_t=info_lattice_t./length(eigvals)

 # plot

 


# Set the figure width (for example, 800 pixels)
fig_width = 800
aspect_ratio = 1.2
fig_height = fig_width / aspect_ratio

# Create the figure
fig = Figure(resolution = (fig_width, fig_height), fontsize = 16)

# Scaling factor
scale_factor = 2 / sqrt(3)

g=1-epsilon
# Create the axis with scaled ticks
ax = Axis(
    fig[1, 1],
    xlabel = "i",
    ylabel = "l",
    xticks = ((0.5:0.5:L) , string.((0.5:0.5:L) )),
    yticks = ((1:L) , string.((1:L) )),
    title = "Info Lattice of avg of eigenstates, L = $L, θ = $theta, g/π = $g",
    )

# Example scatter plot loop
L = 8
for l in 1:L
    for i in 1:(L - l + 1)
        p=info_lattice_t[l][i]
        scatter!(ax, i + (l / 2 )-0.5, l, color = (viridis[27]*(p)+viridis[3]*(1-p)), markersize =20)
    end
end

#fig[1,2]=Colorbar

fig
```
